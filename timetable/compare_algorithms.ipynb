```python
def moving_average(data, window_size):
    smoothed_data = []
    for i in range(len(data)):
        if i < window_size:
            smoothed_data.append(np.mean(data[:i+1]))
        else:
            smoothed_data.append(np.mean(data[i-window_size+1:i+1]))
    return smoothed_data

window_size = 20
```

```python
# Применяем скользящее среднее
smoothed_data = moving_average(times, window_size)

# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(times, label="Оригинальные данные", alpha=0.5)
plt.plot(smoothed_data, label=f"Скользящее среднее (окно={window_size})", color='red', linewidth=2)
plt.legend()
plt.title("Сглаживание графика с помощью скользящего среднего")
plt.show()
```
